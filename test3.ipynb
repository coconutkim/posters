````python
# 훈련셋 준비하기

import os
import random
import shutil

def move_random_images(source_folder, destination_folder, num_images):
    # 원본 폴더에서 이미지 파일 목록 가져오기
    image_extensions = ('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff')
    all_images = [
        f for f in os.listdir(source_folder) 
        if f.lower().endswith(image_extensions) and os.path.isfile(os.path.join(source_folder, f))
    ]
    
    # 선택할 이미지 수가 전체 이미지 수보다 많으면 조정
    # num_images = min(num_images, len(all_images))

    # 무작위로 이미지 선택
    selected_images = random.sample(all_images, num_images)

    # 목적지 폴더가 존재하지 않으면 생성
    os.makedirs(destination_folder, exist_ok=True)

    # 선택된 이미지를 목적지 폴더로 이동
    for image in selected_images:
        src = os.path.join(source_folder, image)
        dst = os.path.join(destination_folder, image)
        shutil.move(src, dst)  # 이동 (복사가 아님)

    print(f"{num_images} images moved to '{destination_folder}'.")

# 경로 설정
folder_path = r'G:\내 드라이브\project_poster\balance_data_rates\#new'
destination_path = r'G:\내 드라이브\project_poster\balance_data_rates\#new_test'

# 함수 호출
move_random_images(folder_path, destination_path, 1750)
````
